```python
import os
import pickle
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Folder yang berisi file pickle
folder_path = r"d:\UNIVERSITAS PERTAMINA\Semester 5\PSD\Tubes\Hasil MFCC\jawa\hamid"

# Mengambil semua file pickle di dalam folder
pickle_files = [f for f in os.listdir(folder_path) if f.endswith('.pkl')]

# Memeriksa apakah ada file pickle
if not pickle_files:
    print("Tidak ada file pickle di folder tersebut.")
else:
    for pickle_file in pickle_files:
        file_path = os.path.join(folder_path, pickle_file)
        
        try:
            # Memuat data MFCC dari file pickle
            with open(file_path, 'rb') as f:
                mfcc_data = pickle.load(f)
        except Exception as e:
            print(f"Error memuat file {pickle_file}: {e}")
            continue

        # Menampilkan struktur data MFCC yang dimuat
        print(f"Memproses {pickle_file}...")

        # Menampilkan heatmap jika data MFCC berbentuk array 2D
        if isinstance(mfcc_data, np.ndarray) and len(mfcc_data.shape) == 2:
            plt.figure(figsize=(10, 6))
            
            # Plot heatmap dengan pembalikan sumbu Y
            sns.heatmap(
                mfcc_data, 
                cmap='viridis', 
                cbar=True,
                yticklabels=np.arange(mfcc_data.shape[0])  # Mengatur sumbu Y mulai dari 0
            )
            
            # Membalikkan sumbu Y
            plt.gca().invert_yaxis()

            plt.title(f"Visualisasi MFCC Heatmap - {pickle_file}")
            plt.xlabel("Frame")
            plt.ylabel("MFCC Coefficients")
            
            # Menampilkan plot
            plt.show()
        else:
            print(f"Data pada {pickle_file} tidak berbentuk array 2D, lewati visualisasi.")

        # Menunggu input untuk melanjutkan ke file berikutnya
        input("Tekan Enter untuk melanjutkan ke file berikutnya...")
```
