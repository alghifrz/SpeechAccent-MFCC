```python
import os
import pickle
import numpy as np

# Mengatur numpy untuk mencetak semua elemen tanpa ringkasan
np.set_printoptions(threshold=np.inf)

# Path folder yang berisi file pickle
folder_path = r"d:\UNIVERSITAS PERTAMINA\Semester 5\PSD\Tubes\Hasil MFCC\medan\diva"

# Menyaring file pickle di folder
pickle_files = [f for f in os.listdir(folder_path) if f.endswith('.pkl')]

# Memuat dan menampilkan isi file pickle satu per satu
for idx, pickle_file in enumerate(pickle_files, start=1):
    file_path = os.path.join(folder_path, pickle_file)
    
    with open(file_path, 'rb') as f:
        data = pickle.load(f)
    
    print(f"=== Iterasi {idx}: File {pickle_file} ===")  # Judul tiap iterasi
    print("Isi matriks:")
    
    # Memeriksa apakah data berbentuk numpy array
    if isinstance(data, np.ndarray):
        print(data)  # Cetak seluruh matriks
    else:
        print("Data bukan numpy array:")
        print(data)  # Cetak data lainnya (jika formatnya berbeda)
    
    print("\n" + "-"*50 + "\n")
```

```python
total_frames = 0
total_files = 0
for idx, pickle_file in enumerate(pickle_files, start=1):
    file_path = os.path.join(folder_path, pickle_file)
    
    with open(file_path, 'rb') as f:
        data = pickle.load(f)
    
    print(f"=== Iterasi {idx}: File {pickle_file} ===")
    if isinstance(data, np.ndarray):
        print(f"Dimensi MFCC: {data.shape}")
        total_frames += data.shape[1]
        total_files += 1

if total_files > 0:
    avg_frames = total_frames / total_files
    print(f"\nJumlah file: {total_files}")
    print(f"Rata-rata jumlah frame (dimensi kedua MFCC): {avg_frames:.2f}")
    print(f"Dimensi MFCC rata rata: ({data.shape[0]}, {avg_frames})")
```
